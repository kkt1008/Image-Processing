```python
img_lena = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Image_Processing_2025_1/lena.bmp', cv2.IMREAD_GRAYSCALE)
plt.title('original')
plt.imshow(img_lena,cmap="gray")
```

```python
#이미지 회전
# 시계방향 90도 회전
img_rotate_90=cv2.rotate(img_lena, cv2.ROTATE_90_CLOCKWISE)
plt.title('rotate 90 clockwise')
plt.imshow(img_rotate_90,cmap="gray")
plt.show()

# 180도 회전
img_rotate_180=cv2.rotate(img_lena, cv2.ROTATE_180)
plt.title('rotate 180 clockwise')
plt.imshow(img_rotate_180,cmap="gray")
plt.show()

# 시계방향 270도 회전 = 반시계방향 90도 회전
img_rotate_270=cv2.rotate(img_lena, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.title('rotate 270 clockwise')
plt.imshow(img_rotate_270,cmap="gray")
plt.show()
```

```python
# BGR 이미지를 RGB로 변환해서 출력
# opencv는 기본적으로 BGR 형식으로 이미지를 불러오기 때문
img_lena = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Image_Processing_2025_1/lena.bmp') # 흑백형식이 아닌 BGR 형식으로 불러온다
img_lena_rgb = cv2.cvtColor(img_lena, cv2.COLOR_BGR2RGB)

plt.title('RGB image')
plt.imshow(img_lena_rgb)
plt.show()

```
