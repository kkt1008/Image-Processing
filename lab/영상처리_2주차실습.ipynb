```python
### 2번 Design a function to convert RGB color image to gray
### Y = 0.3R + 0.5G + 0.2B 이미지의 RGB비율을 조절하여 새로운 이미지를 만듬
### rgb_lena이미지가 ndarray이므로 행렬의 각 차원별로 조절 가능

R=lena_img[:,:,0] #2차원 정방행렬
G=lena_img[:,:,1] #2차원 정방행렬
B=lena_img[:,:,2] #2차원 정방행렬
#Y= 0.3*R+0.5*G+0.2*B # 2차원 이미지 BGR 형태
new_R = R*0.3
new_G = G*0.5
new_B = B*0.2
new_img = new_R + new_G + new_B # data type은 64비트
new_cvt_lena_img = np.clip(new_img,0,255)
new_cvt_lena_img = new_cvt_lena_img.astype('uint8')

# cv2.cvtColor()함수는 rgb로 변환시 8비트 데이터만 변환 가능 (2의 8승이 256이기 때문에 0~255개의 화솟값 존재)
new_cvt_rgb_lena_img = cv2.cvtColor(new_cvt_lena_img,cv2.COLOR_GRAY2RGB)
plt.imshow(new_cvt_rgb_lena_img) # 최종 이미지 회색으로 출
```

```python
### 6번 Design a function to find statistics of given image
### input:image / output:mean, variance, max, mim

def image_statistics(img):
    mean=np.mean(img)
    variance=np.var(img)
    max=np.max(img)
    min=np.min(img)
    return mean, variance, max, min

mean_val,variance_val,max_val,min_val=image_statistics(rgb_lena_img)
print("mean : ", mean_val)
print("variance : ",variance_val)
print("max : ",max_val)
print("min : ",min_val)
```

```python
#2주차 개인 공부용 코드--이미지 히스토그램 그리기(흑백)

# 흑백이미지 히스토그램
def compute_histogram_gray(image):
    hist = np.zeros(256)  # 픽셀 값(0~255)에 대한 빈도수 저장

    # 픽셀 개수 세기
    for row in image:
        for pixel in row:
            hist[pixel] += 1

    return hist

hist_gray_img = compute_histogram_gray(gray_lena_img)

# 흑백 이미지 히스토그램 그리기
plt.plot(hist_gray_img, color='black')
plt.title("Grayscale Image Histogram")
plt.xlabel("Pixel Value (0-255)")
plt.ylabel("Frequency")
plt.xlim([0, 256])
plt.grid()
plt.show()
```

```python
#2주차 개인 공부용 코드--이미지 히스토그램 그리기(컬러)

def compute_histogram_rgb(image):
    """RGB 이미지의 히스토그램을 직접 계산"""
    hist_r = np.zeros(256)
    hist_g = np.zeros(256)
    hist_b = np.zeros(256)

    # 이미지의 각 픽셀을 순회하면서 값 세기
    for row in image:
        for pixel in row:
            hist_r[pixel[0]] += 1  # Red 채널
            hist_g[pixel[1]] += 1  # Green 채널
            hist_b[pixel[2]] += 1  # Blue 채널

    return hist_r, hist_g, hist_b

hist_r, hist_g, hist_b = compute_histogram_rgb(rgb_lena_img)

# 컬러 이미지 히스토그램 그리기
plt.plot(hist_r, color='r', label='Red')
plt.plot(hist_g, color='g', label='Green')
plt.plot(hist_b, color='b', label='Blue')
plt.title("RGB Image Histogram")
plt.xlabel("Pixel Value (0-255)")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
plt.show()
```
